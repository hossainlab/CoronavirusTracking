# Tracking the Spread of Coronavirus?
**Coronaviruses** are a family of viruses known for containing strains that cause potentially deadly diseases in mammals and birds. In humans they're typically spread via airborne droplets of fluid produced by infected individuals.Some rare but notable strains, including Wuhan coronavirus (2019-nCoV), and those responsible for severe acute respiratory syndrome (SARS) and Middle East respiratory syndrome (MERS), can cause death in humans.[See More](https://www.sciencealert.com/coronavirus)

## Coronavirus Dataset
The 2019-nCoV is a contagious coronavirus that hailed from Wuhan, China. This new strain of virus has striked fear in many countries as cities are quarantined and hospitals are overcrowded. This dataset will help us understand how 2019-nCoV is spread aroud the world.

[See More](https://www.kaggle.com/brendaso/2019-coronavirus-dataset-01212020-01262020#2019_nC0v_20200121_20200126%20-%20SUMMARY.csv)

## About this dataset
This file is an aggregated version of the 2019-nCoV dataset collected by Johns Hopkins University.

## Data Grouping

```python
# Group by DateTime 
data_by_date = cv_data.groupby('Last Update').sum().reset_index()
data_by_date
```

```python
# Group by country 
data_by_country = cv_data.groupby('Country').agg('sum').reset_index()
data_by_country
```

```python
# Group by State 
data_by_state = cv_data.groupby('Province/State').agg('sum').reset_index()
data_by_state
```

## Count Plot

```python
def count_plot(feature, value, title, df, size=1):
    """
    Take feature, value, title, df, size as inputs.And returns the bar plot of diffrent kinds of DataFrame
    """
    fig, ax = plt.subplots(1,1, figsize=(4*size,4))
    g = sns.barplot(df[feature], df[value],  palette='Set3')
    g.set_title("Number of {}".format(title))
    ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
    plt.show() 
```

```python
# Number of confirmed cases by Province/Satate
count_plot('Province/State', 'Confirmed', 'Confirmed Cases(By State)', data_by_state, size=4)
```

```python
# Number of recovered cases by Province/State
count_plot('Province/State', 'Recovered', 'Recovered Cases(By State)', data_by_state, size=4)
```

```python
# Number of Deaths cases by Province/State
count_plot('Province/State', 'Suspected', 'Suspected Cases(By State)', data_by_state, size=4)
```

```python
# Number of Deaths cases by Province/State
count_plot('Province/State', 'Deaths', 'Deaths Cases(By State)', data_by_state, size=4)
```

```python
# Number of confirmed cases by DateTime
data_by_date.sort_values(by='Last Update', inplace=True, ascending=False)
count_plot('Last Update', 'Confirmed', 'Confirmed Cases(DateTime)', data_by_date, size=4)
```

```python
# Number of recovered cases by DateTime
data_by_date.sort_values(by='Last Update', inplace=True, ascending=False)
count_plot('Last Update', 'Recovered', 'Recovered Cases(DateTime)', data_by_date, size=4)
```

```python
# Number of deaths cases by DateTime
data_by_date.sort_values(by='Last Update', inplace=True, ascending=False)
count_plot('Last Update', 'Deaths', 'Deaths Cases(DateTime)', data_by_date, size=4)
```

```python
# Number of suspected cases by DateTime
data_by_date.sort_values(by='Last Update', inplace=True, ascending=False)
count_plot('Last Update', 'Suspected', 'Suspected Cases(DateTime)', data_by_date, size=4)
```

```python
# Number of confirmed cases by Country
count_plot('Country', 'Confirmed', 'Confirmed Cases(Country Wise)', data_by_country, size=4)
```

## Time Variation

```python
def plot_time_variation(df, y='Confirmed', hue='Province/State', size=1):
    fig, ax = plt.subplots(1,1, figsize=(4*size,3*size))
    g = sns.lineplot(x="Last Update", y=y, hue=hue, data=cv_data)
    plt.xticks(rotation=90)
    plt.title(f'{y} Cases Grouped by {hue}')
    plt.show() 
```

```python
# Time variation of confirmed cases grouped by Province/State 
plot_time_variation(data_by_date, size=4)
```

```python
# Time variation of recovered cases grouped by Province/State 
plot_time_variation(data_by_date, y='Recovered', size=4)
```

```python
# Time variation of death cases grouped by Province/State 
plot_time_variation(data_by_date, y='Deaths', size=4)
```

```python
# Time variation of suspected cases grouped by Province/State 
plot_time_variation(data_by_date, y='Suspected', size=4)
```

## States of China

```python
# States
china_data['Province/State']
```

```python
# Number of confirmed cases by States(in China)
china_data.sort_values(by='Confirmed', inplace=True, ascending=False)
count_plot('Province/State', 'Confirmed', 'Patients Confirmed Infected by Corona Virus(By States)'
, china_data, size=4)
```

```python
# Number of recovered cases by States(in China)
china_data.sort_values(by='Recovered', inplace=True, ascending=False)
count_plot('Province/State', 'Recovered', 'Recovered Cases in China(By States)'
, china_data, size=4)
```

```python
# Number of deaths cases by States(in China)
china_data.sort_values(by='Deaths', inplace=True, ascending=False)
count_plot('Province/State', 'Deaths', 'Deaths Cases in China(By States)'
, china_data, size=4)
```

```python
# Time variation of confirmed cases in China grouped by Province/State 
plot_time_variation(china_data, size=4)
```

```python
# Time variation of recovered cases in China grouped by Province/State 
plot_time_variation(china_data, y='Recovered', size=4)
```
